```python
import pandas as pd
import numpy as np
```

```python
final_input_data = pd.read_csv('C:/Users/patel/Thesis/DataCleaning/BoP/inputdata.csv')
clothing_expense = pd.read_csv('Clothing_TotalCost.csv')
unit_clothing_expense = pd.read_csv('FinalUnitcosts.csv')
```

```python
clothing_expense = clothing_expense[clothing_expense['UNIT']=='Current prices, million euro']
clothing_expense.head()
clothing_expense['Value'] = clothing_expense['Value'].replace({' ':''},regex=True)
clothing_expense['Value'] = pd.to_numeric(clothing_expense['Value'],errors='coerce')
clothing_expense = clothing_expense[['TIME','GEO','Value']]
clothing_expense.replace({":":np.NaN}, inplace=True)
clothing_expense.head()
```

```python
unit_clothing_expense['Average per piece'] = unit_clothing_expense.iloc[:,2:].mean(axis=1)
```

```python
unit_clothing_expense = unit_clothing_expense[['TIME','GEO','Average per piece']]
final_clothing_expenses = pd.merge(clothing_expense,unit_clothing_expense,how='inner',on=['TIME','GEO'])
final_clothing_expenses['Total units'] = (final_clothing_expenses['Value']/final_clothing_expenses['Average per piece'])*(1000000)
final_clothing_consumption = final_clothing_expenses[['TIME','GEO','Total units']]
final_data = pd.merge(final_input_data,final_clothing_consumption,how='inner',on=['TIME','GEO'])
final_data['consumption_per_capita'] = final_data['Total units']/final_data['Total_pop']
final_data.drop(['Total units', 'Total_pop','TIME','GEO'], axis=1, inplace=True)
final_data.to_csv('final_input_output.csv',index=False)
final_data.head()
```
