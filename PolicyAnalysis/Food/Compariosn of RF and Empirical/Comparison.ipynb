```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
empirical_data = pd.read_csv('Foodconsumedperbuurt.csv')
```

## Comparison of potatoes consumption

```python
food_name = '01. Potatoes and other tubers'
food_consumed = empirical_data[empirical_data['food']==food_name]
food_consumed = food_consumed[food_consumed['Type']=='Buurt']
food_consumed['Amount per year'] = food_consumed['Amount per day']*365/1000
food_consumed = food_consumed[['Name','Amount per year']]
rf_food_consumed = pd.read_csv('PotatoRF.csv')
rf_food_consumed = rf_food_consumed[['Name','Total Consumption']]
food_comparison = pd.merge(food_consumed,rf_food_consumed,how='inner',on=['Name'])
food_comparison.head()
plt.figure(figsize=(12,6))
ax=plt.axes()
plt.scatter(food_comparison['Amount per year'],food_comparison['Total Consumption'],color='g')
plt.xlabel('Empirical value',fontsize=15)
plt.ylabel('Predicted value_rf',fontsize=15)
#plt.ylim(5,30)
plt.xlim(0,food_comparison['Total Consumption'].max())
x = np.linspace(0,food_comparison['Total Consumption'].max(),100)
y = x
plt.plot(x,y,'-.r',label='y=x')
plt.title('Potato Consumption',fontsize=20)
```
