```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
buurt_data = pd.read_excel('Hague_cijfers.xlsx',sheet_name='Meest-recent')
```

```python
buurt_data.rename(columns = {'Soort regio':'type','Naam van het gebied':'Name'}, inplace = True) 
```

```python
nl_tert_25to34 = 0.41
nl_tert_55to64 = 0.27
nl_tert_ratio_youngtoold = nl_tert_25to34/nl_tert_55to64
```

```python
#buurt_data=buurt_data.rename(columns = {'Soort_regio':'type','Naam van het gebied':'Name'}, inplace = True) 
wijk = ''
buurt_data['Nearest wijk'] = 'Den Haag'
for i in range (0,len(buurt_data)):
    if (buurt_data.loc[i,'type'] == 'Wijk'):
        wijk = buurt_data.loc[i,'Name']
    buurt_data.loc[i,'Nearest wijk'] = wijk
buurt_data['Wijk no'] = buurt_data['Nearest wijk'].str.split(" ",n=2,expand=True)[1]
```

```python
hague_primary_rate = '33'
hague_secondary_rate = '36'
hague_tertiary_rate = '31'
education = pd.read_csv('Education.csv')
education['Wijk no'] = education['Wijken'].str.split(" ",n = 1, expand = True)[0]
buurt_data_with_edu = buurt_data.merge(education,on=['Wijk no'])
```

```python
buurt_data_needed = buurt_data_with_edu[['Name','type','Codering','Aantal inwoners','Mannen',
                                       'Vrouwen','0 tot 15 jaar','15 tot 25 jaar','25 tot 45 jaar', '45 tot 65 jaar',
                                       '65 jaar of ouder','Wijken',
                                       'Opleidingsniveau Laag|2017', 'Opleidingsniveau Middelbaar|2017',
                                       'Opleidingsniveau Hoog|2017']]
buurt_data_needed.columns =['Name','Type','Code','Population','Man','Woman','0 to 15 year','15 to 25 year','25 to 45 year',
                            '45 to 65 year','65 year or older','Wijken',
                           'Primary education','Secondary education','Tertiary education']
```

```python
buurt_data_needed['Primary education'] = buurt_data_needed['Primary education'].fillna(hague_primary_rate)
buurt_data_needed['Secondary education'] = buurt_data_needed['Secondary education'].fillna(hague_secondary_rate)
buurt_data_needed['Tertiary education'] = buurt_data_needed['Tertiary education'].fillna(hague_tertiary_rate)
buurt_data_needed['Primary education'].replace({"x":hague_primary_rate}, inplace=True)
buurt_data_needed['Secondary education'].replace({"x":hague_secondary_rate}, inplace=True)
buurt_data_needed['Tertiary education'].replace({"x":hague_tertiary_rate}, inplace=True)
```

```python
buurt_data_needed['Primary education'] = 0.01*buurt_data_needed['Primary education'].astype('float')
buurt_data_needed['Secondary education'] = 0.01*buurt_data_needed['Secondary education'].astype('float')
buurt_data_needed['Tertiary education'] = 0.01*buurt_data_needed['Tertiary education'].astype('float')
buurt_data_needed['60 to 70 year'] = buurt_data_needed['45 to 65 year']*(1/4) + buurt_data_needed['65 year or older']*(5/17)
buurt_data_needed['70 year or older'] = buurt_data_needed['65 year or older']*(12/17) 
buurt_data_needed['less than 40'] =  buurt_data_needed['0 to 15 year']+ buurt_data_needed['15 to 25 year'] + (3/4)*buurt_data_needed['25 to 45 year']
buurt_data_needed['more than 40'] = buurt_data_needed['Population'] - buurt_data_needed['less than 40']
buurt_data_needed['old_literacy_rate'] = (buurt_data_needed['Tertiary education']*buurt_data_needed['Population'])/(buurt_data_needed['more than 40'] + nl_tert_ratio_youngtoold*(buurt_data_needed['less than 40']))
buurt_data_needed['young_literacy_rate'] = buurt_data_needed['old_literacy_rate'] *nl_tert_ratio_youngtoold
buurt_data_needed['40 to 60 year'] = buurt_data_needed['Population']-buurt_data_needed['less than 40']-buurt_data_needed['60 to 70 year']- buurt_data_needed['70 year or older']
```

```python
buurt_data_needed['male less than 40'] = buurt_data_needed['less than 40']*buurt_data_needed['Man']/buurt_data_needed['Population']
buurt_data_needed['female less than 40'] = buurt_data_needed['less than 40']*buurt_data_needed['Woman']/buurt_data_needed['Population']
buurt_data_needed['male 40 to 60'] = buurt_data_needed['40 to 60 year']*buurt_data_needed['Man']/buurt_data_needed['Population']
buurt_data_needed['female 40 to 60'] = buurt_data_needed['40 to 60 year']*buurt_data_needed['Woman']/buurt_data_needed['Population']
buurt_data_needed['male 60 to 70'] = buurt_data_needed['60 to 70 year']*buurt_data_needed['Man']/buurt_data_needed['Population']
buurt_data_needed['female 60 to 70'] = buurt_data_needed['60 to 70 year']*buurt_data_needed['Woman']/buurt_data_needed['Population']
buurt_data_needed['male more than 70'] = buurt_data_needed['70 year or older']*buurt_data_needed['Man']/buurt_data_needed['Population']
buurt_data_needed['female more than 70'] = buurt_data_needed['70 year or older']*buurt_data_needed['Woman']/buurt_data_needed['Population']
buurt_data_needed['male 40 to 60'] = buurt_data_needed['40 to 60 year']*buurt_data_needed['Man']/buurt_data_needed['Population']
buurt_data_needed['female 40 to 60'] = buurt_data_needed['40 to 60 year']*buurt_data_needed['Woman']/buurt_data_needed['Population']
buurt_data_needed['male 60 to 70'] = buurt_data_needed['60 to 70 year']*buurt_data_needed['Man']/buurt_data_needed['Population']
buurt_data_needed['female 60 to 70'] = buurt_data_needed['60 to 70 year']*buurt_data_needed['Woman']/buurt_data_needed['Population']
buurt_data_needed['male more than 70'] = buurt_data_needed['70 year or older']*buurt_data_needed['Man']/buurt_data_needed['Population']
buurt_data_needed['female more than 70'] = buurt_data_needed['70 year or older']*buurt_data_needed['Woman']/buurt_data_needed['Population']
```

```python
buurt_data_needed['male less than 40 educated'] = buurt_data_needed['male less than 40']*buurt_data_needed['young_literacy_rate']
buurt_data_needed['male less than 40 uneducated'] = buurt_data_needed['male less than 40']*(1-buurt_data_needed['young_literacy_rate'])
buurt_data_needed['female less than 40 educated'] = buurt_data_needed['female less than 40']*buurt_data_needed['young_literacy_rate']
buurt_data_needed['female less than 40 uneducated'] = buurt_data_needed['female less than 40']*(1-buurt_data_needed['young_literacy_rate'])
buurt_data_needed['male 40 to 60 educated'] = buurt_data_needed['male 40 to 60']*buurt_data_needed['old_literacy_rate']
buurt_data_needed['male 40 to 60 uneducated'] = buurt_data_needed['male 40 to 60']*(1-buurt_data_needed['old_literacy_rate'])
buurt_data_needed['female 40 to 60 educated'] = buurt_data_needed['female 40 to 60']*buurt_data_needed['old_literacy_rate']
buurt_data_needed['female 40 to 60 uneducated'] = buurt_data_needed['female 40 to 60']*(1-buurt_data_needed['old_literacy_rate'])
buurt_data_needed['male 60 to 70 educated'] = buurt_data_needed['male 60 to 70']*buurt_data_needed['old_literacy_rate']
buurt_data_needed['male 60 to 70 uneducated'] = buurt_data_needed['male 60 to 70']*(1-buurt_data_needed['old_literacy_rate'])
buurt_data_needed['female 60 to 70 educated'] = buurt_data_needed['female 60 to 70']*buurt_data_needed['old_literacy_rate']
buurt_data_needed['female 60 to 70 uneducated'] = buurt_data_needed['female 60 to 70']*(1-buurt_data_needed['old_literacy_rate'])
buurt_data_needed['male more than 70 educated'] = buurt_data_needed['male more than 70']*buurt_data_needed['old_literacy_rate']
buurt_data_needed['male more than 70 uneducated'] = buurt_data_needed['male more than 70']*(1-buurt_data_needed['old_literacy_rate'])
buurt_data_needed['female more than 70 educated'] = buurt_data_needed['female more than 70']*buurt_data_needed['old_literacy_rate']
buurt_data_needed['female more than 70 uneducated'] = buurt_data_needed['female more than 70']*(1-buurt_data_needed['old_literacy_rate'])
buurt_data_needed['male less than 60 educated'] = buurt_data_needed['male less than 40 educated'] + buurt_data_needed['male 40 to 60 educated']
buurt_data_needed['male less than 60 uneducated'] = buurt_data_needed['male less than 40 uneducated'] + buurt_data_needed['male 40 to 60 uneducated']
buurt_data_needed['female less than 60 educated'] = buurt_data_needed['female less than 40 educated'] + buurt_data_needed['female 40 to 60 educated']
buurt_data_needed['female less than 60 uneducated'] = buurt_data_needed['female less than 40 uneducated'] + buurt_data_needed['female 40 to 60 uneducated']
```

```python
buurt_data['income per household'] = buurt_data['Gemiddeld inkomen per inkomensontvanger ']*buurt_data['Gemiddelde huishoudensgrootte']/12
buurt_data_needed['income per household'] = np.array(buurt_data['income per household'])[1:159]
```

```python
sd_malerespondent = 0.4986
sd_age = 13.62
sd_collegeeducation = 0.4879
sd_income = 4684
sd_greenpower = 0.4997
coeff_male = -0.032*sd_greenpower/sd_malerespondent
coeff_age_morethansixty = -0.243*sd_greenpower*65/sd_age
coeff_age_morethanseventy = -0.067*sd_greenpower*77/sd_age
coeff_income = 0.017*sd_greenpower/sd_income
coeff_edu = 0.159*sd_greenpower/sd_collegeeducation
constant = 0.059
for i in range (0,len(buurt_data_needed)):
    income = buurt_data_needed.loc[i,'income per household']
#    income = 0
    buurt_data_needed.loc[i,'prob_male less than 60 educated'] = 1/(1 + math.exp(-1*(coeff_male+coeff_edu + constant+ coeff_income*income)))
    buurt_data_needed.loc[i,'prob_male less than 60 uneducated'] = 1/(1 + math.exp(-1*(coeff_male+ constant + coeff_income*income)))
    buurt_data_needed.loc[i,'prob_female less than 60 educated'] = 1/(1 + math.exp(-1*(coeff_edu + constant + coeff_income*income)))
    buurt_data_needed.loc[i,'prob_female less than 60 uneducated'] = 1/(1 + math.exp(-1*(constant+ coeff_edu*income)))
    buurt_data_needed.loc[i,'prob_male 60 to 70 educated'] = 1/(1 + math.exp(-1*(coeff_male+coeff_edu +constant+ coeff_income*income + coeff_age_morethansixty)))
    buurt_data_needed.loc[i,'prob_male 60 to 70 uneducated'] = 1/(1 + math.exp(-1*(coeff_male + constant+ coeff_income*income + coeff_age_morethansixty)))
    buurt_data_needed.loc[i,'prob_female 60 to 70 educated'] = 1/(1 + math.exp(-1*(coeff_edu + constant+ coeff_income*income+coeff_age_morethansixty)))
    buurt_data_needed.loc[i,'prob_female 60 to 70 uneducated'] = 1/(1 + math.exp(-1*( constant+ coeff_income*income+coeff_age_morethansixty)))
    buurt_data_needed.loc[i,'prob_male more than 70 educated'] = 1/(1 + math.exp(-1*(coeff_male+ coeff_edu + constant+ coeff_income*income+coeff_age_morethanseventy)))
    buurt_data_needed.loc[i,'prob_male more than 70 uneducated'] = 1/(1 + math.exp(-1*(coeff_male + constant+ coeff_income*income+coeff_age_morethanseventy)))
    buurt_data_needed.loc[i,'prob_female more than 70 educated'] = 1/(1 + math.exp(-1*(coeff_edu + constant + coeff_income*income+coeff_age_morethanseventy)))
    buurt_data_needed.loc[i,'prob_female more than 70 uneducated'] = 1/(1 + math.exp(-1*(constant+ coeff_income*income+coeff_age_morethanseventy)))
```

```python
ppl_clusters = ['male less than 60 educated','male less than 60 uneducated', 'female less than 60 educated',
                'female less than 60 uneducated', 'male 60 to 70 educated', 'male 60 to 70 uneducated',
                'female 60 to 70 educated', 'female 60 to 70 uneducated', 'male more than 70 educated',
                'male more than 70 uneducated', 'female more than 70 educated','female more than 70 uneducated'] 

probabilities = ['prob_male less than 60 educated', 'prob_male less than 60 uneducated','prob_female less than 60 educated',
                 'prob_female less than 60 uneducated', 'prob_male 60 to 70 educated','prob_male 60 to 70 uneducated',
                 'prob_female 60 to 70 educated','prob_female 60 to 70 uneducated', 'prob_male more than 70 educated',
                 'prob_male more than 70 uneducated','prob_female more than 70 educated','prob_female more than 70 uneducated']
for i in range (0,len(buurt_data_needed)):
    probability = 0
    for j in range (0,len(ppl_clusters)):
        probability = probability + (buurt_data_needed.loc[i,ppl_clusters[j]]*buurt_data_needed.loc[i,probabilities[j]])
    probability = probability/buurt_data_needed.loc[i,'Population']
    buurt_data_needed.loc[i,'Net probability'] = probability
```

```python
buurt_data_needed = buurt_data_needed[['Name','Type','Population','Net probability']]
buurt_data = buurt_data[['Name','Huishoudens totaal','Gemiddeld elektriciteitsverbruik totaal','Gemiddeld aardgasverbruik totaal']]
energy_consumption = pd.merge(buurt_data,buurt_data_needed,on=['Name'],how='inner')
```

```python
energy_consumption.head()
```

```python
energy_consumption = energy_consumption.dropna()
energy_consumption.columns = ['Name','Total_households','Average electricity consumption','Average gas consumption','Type','Population','Probability']
energy_consumption['green electricity'] = energy_consumption['Average electricity consumption']*energy_consumption['Probability']*energy_consumption['Total_households']/energy_consumption['Population']
energy_consumption['nongreen electricity'] = energy_consumption['Average electricity consumption']*(1-energy_consumption['Probability'])*energy_consumption['Total_households']/energy_consumption['Population']
energy_consumption['green gas'] = energy_consumption['Average gas consumption']*energy_consumption['Probability']*energy_consumption['Total_households']/energy_consumption['Population']
energy_consumption['nongreen gas'] = energy_consumption['Average gas consumption']*(1-energy_consumption['Probability'])*energy_consumption['Total_households']/energy_consumption['Population']
final_gas = energy_consumption[['Name','Type','green gas','nongreen gas']]
final_electricity =energy_consumption[['Name','Type','green electricity','nongreen electricity']]
final_gas.to_csv('final_gas_consumption.csv',index=False)
final_electricity.to_csv('final_electricity_consumption.csv',index=False)
```
